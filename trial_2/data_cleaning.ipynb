# Cleaning Data

```python
import pandas as pd
import numpy as np

data_path = "./data/train.csv"

data = pd.read_csv(data_path)
```

```python
data.head()
```

```python
data.describe()
```

```python
len(data)
```

```python
from sklearn.impute import SimpleImputer

# using an imputer to deal with empty spaces
imputer = SimpleImputer(strategy="most_frequent")

cleaned_data = pd.DataFrame(imputer.fit_transform(data), columns=data.columns)
```

```python
cleaned_data.head()
```

```python
import os

def save_cleaned_data(data):
    # a function for saving the cleaned daat
    directory = os.path.join("data", "cleaned")
    
    os.makedirs(directory, exist_ok=True)

    data.to_csv(os.path.join(directory, "cleaned_data.csv"), index=False)
```

```python
# saving the imputed data
save_cleaned_data(cleaned_data)
```
