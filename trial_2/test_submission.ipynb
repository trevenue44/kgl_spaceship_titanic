```python
import numpy as np
import pandas as pd

test = pd.read_csv("./data/test.csv")
```

```python
test.head()
```

```python
# selecting all the useful features of the X

from sklearn.impute import SimpleImputer

# using an imputer to deal with empty spaces
imputer = SimpleImputer(strategy="most_frequent")

test = pd.DataFrame(imputer.fit_transform(test), columns=test.columns)

# we don't use the name column
X = test.drop("Name", axis=1)
passenger_id = X["PassengerId"]
X["PassengerGroup"] = passenger_id.apply(lambda x: int(x[:4]))
X = X.drop("PassengerId", axis=1)
```

```python
# loading the model
import pickle

with open("./saved_models/model_1.pkl", "rb") as f:
    model = pickle.load(f)
```

```python
X.describe()
```

```python
predicted = model.predict(X)
```

```python
transported = predicted.astype(np.bool8)
```

```python
result = pd.DataFrame({"PassengerId": test["PassengerId"], "Transported": transported})
```

```python
result.head()
```

```python
import os

os.makedirs("submissions", exist_ok=True)

result.to_csv("./submissions/submission_1.csv", index=False)
```
